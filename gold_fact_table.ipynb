```python

df_silver = spark.sql("SELECT * FROM parquet.`abfss://silver@cargoutdatalake.dfs.core.windows.net/carsales`")
```

```python
df_silver.display()
```

```python
df_model = spark.sql("select * from car_project_gout.gold.dim_model1")
df_branch = spark.sql("select * from car_project_gout.gold.dim_branch")
df_dealer = spark.sql("select * from car_project_gout.gold.dim_dealer")
df_date = spark.sql("select * from car_project_gout.gold.dim_date")
```

```python
df_fact = df_silver \
    .join(df_model, df_silver.Model_ID == df_model.model_id, "left") \
    .join(df_branch, df_silver.Branch_ID == df_branch.branch_id, "left") \
    .join(df_dealer, df_silver.Dealer_ID == df_dealer.dealer_id, "left") \
    .join(df_date, df_silver.Date_ID == df_date.date_id, "left") \
    .select(
        df_silver.Revenue, 
        df_silver.Units_Sold,  # ✅ Add `units_sold`
        df_model.dim_model_key, 
        df_branch.dim_branch_key, 
        df_dealer.dim_dealer_key, 
        df_date.dim_date_key
    )

from pyspark.sql.functions import sum

df_fact = df_fact \
    .groupBy("dim_model_key", "dim_branch_key", "dim_dealer_key", "dim_date_key") \
    .agg(
        sum("Revenue").alias("Revenue"),  # Sum revenue to handle duplicates
        sum("Units_Sold").alias("Units_Sold")  # Sum units sold (if applicable)
    )
```

```python
df_fact.display()
```

```python
from delta.tables import DeltaTable


if spark.catalog.tableExists("car_project_gout.gold.fact_sales"):
    dtl_Table = DeltaTable.forName(spark, "car_project_gout.gold.fact_sales")
    
    dtl_Table.alias("dtl").merge(
        df_fact.alias("df"),
        "dtl.dim_model_key = df.dim_model_key AND \
         dtl.dim_branch_key = df.dim_branch_key AND \
         dtl.dim_dealer_key = df.dim_dealer_key AND \
         dtl.dim_date_key = df.dim_date_key"
    ).whenMatchedUpdateAll().whenNotMatchedInsertAll().execute()
else:
    df_fact.write.format("delta").mode("overwrite").option(
        'path', 'abfss://gold@cargoutdatalake.dfs.core.windows.net/fact_sales'
    ).saveAsTable("car_project_gout.gold.fact_sales")
```
